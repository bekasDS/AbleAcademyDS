### 1. Блок импорта

```python
# константы
from listutils import (fake_file_schools,       # школы
                       fake_file_humans,        # люди (ученики и учителя)
                       fake_file_nation,        # национальность
                       fake_file_subjects,      # предметы
                       fakesrcdir               # путь записи файла
                       )
# фейковые генераторы данных
from faker import Faker
from faker_education import SchoolProvider
# генератор чисел
import random
# системные  утилиты, работа с путями
from os import makedirs
from os.path import isdir
# работа с данными
import json

# инициализация фейковых данных
fake = Faker('ru_RU')
# fake = Faker()
fake.add_provider(SchoolProvider)
Faker.seed(0)   # '0' позволяет сохранять генерируемые данные при определённых условиях
```

### 2. <span style="color: red;">КОНСТАНТЫ</span>
В этом блоке можно вносить изменения.

```python
# КОНСТАНТЫ

# путь записываемых файлов
mydir = '../../' + fakesrcdir

# количество /итераций/проходов/строк/позиций
it_schools: int = 3      # школы
it_humans:int   = 10000  # люди
it_nation:int   = 255    # страны

# количество людей ОТ/ДО
# ученики
num_stud_1:int = 50
num_stud_2:int = 150
# учителя
num_tch_1:int  = 5
num_tch_2:int  = 15

# число школ ОТ/ДО
num_sch_1 = 1
num_sch_2 = 200

# название генерации учебного заведения
sname = "School №"
```

### 3. Служебные функции

# Генерация данных в файлы:

<code> на диске создаются 2 файла со списком школ и людей </code>

#### **1.** Заполнение данных по ШКОЛАМ в файл

```python
# ЗАПОЛНЕНИЕ ДАННЫХ ПО ШКОЛАМ В ФАЙЛ

school_set = set()    # названия школ
my_school_dict = {}   # данные по школам

# создание НАЗВАНИЯ ШКОЛ
for i in range(it_schools):
    # заполнение словаря строками данных по людям
    school_set.add(sname + str(random.randint(num_sch_1, num_sch_2)))

i = 0
# создание ДАННЫХ по школам
for school in school_set:
    my_school_dict[i] = {'name': school,                                              # название школы
                         'address': fake.street_address(),                            # адрес
                         'phone': fake.phone_number().replace(" ", ""),               # телефон
                         'email': fake.email(),                                       # электронная почта
                         'students_count': random.randrange(num_stud_1, num_stud_2),  # количество учеников
                         'teachers_count': random.randrange(num_tch_1, num_tch_2)     # количество учителей
                    }
    i += 1

# запись в файл
# создание каталога
if not isdir(mydir): makedirs(mydir)

with open(mydir + "/" + fake_file_schools, "w", encoding="UTF-8", newline="") as file:
    # for val in my_school_dict:
    #     file.writelines(str(val) + '\n')
    
    json.dump(my_school_dict, file, indent=3, ensure_ascii=False)

    print(f"{fake_file_schools} создан")

print(school_set)
print(my_school_dict)
```

#### **2.** Заполнение данных по ЛЮДЯМ в файл

```python
# ЗАПОЛНЕНИЕ ДАННЫХ ПО ЛЮДЯМ В ФАЙЛ
mydict = []  # данные по людям

mydict2 = []

for i in range(it_humans):
    # заполнение словаря строками данных по людям
    mydict.append(fake.simple_profile())

    st = str(fake.simple_profile())
    st = st.replace(r"'", r'"').replace('\\', '') \
           .replace('datetime', '"datetime') \
           .replace(')}', ')"}').replace("'", "\"")

    mydict2.append(st.strip('"'))

    # ОТЛАДОЧНАЯ ИНФОРМАЦИЯ
    # **********************************

    # mydict2.append("{" + st + "}")

    # mydict2.append(str(fake.simple_profile()).replace(r"'", r'"').replace('\\', '')).replace('datetime', '"datetime').replace(')}', ')"}')
    
    # print(mydict2[i])

    # print(json.loads(fake.simple_profile()))
    # print(json.loads(str(fake.simple_profile()).replace("'", "\"")))
    # exit()
    # **********************************


# запись в файл
# создание каталога
if not isdir(mydir): makedirs(mydir)

with open(mydir + "/" + fake_file_humans, "w", encoding="UTF-8", newline="") as file:
    file.writelines("[\n")
    i = 1
    for val in mydict2:
        sep = "," if (i < len(mydict2)) else ""
        file.writelines(str(val + sep).replace("'", "\"") + '\n')
        i += 1
    file.writelines("]\n")

    print(f"{fake_file_humans} создан")


# # запись в файл
# # создание каталога
# if not isdir(mydir): makedirs(mydir)

# with open(mydir + "/" + fake_file_humans, "a+", encoding="UTF-8", newline="") as file:
#     for val in mydict:
#         file.writelines(str(val) + '\n')

#     print(f"{fake_file_humans} создан")
```

### **3.** Национальность
берётся страна

```python
# ЗАПОЛНЕНИЕ ДАННЫХ ПО СТРАНАМ В ФАЙЛ
mydict = set()  # данные по странам

mydict2 = {}

for i in range(it_nation):
    # заполнение словаря строками данных по странам
    mydict.add(fake.country_code().replace(" ",""))

    st = str(mydict)

    # удаляем лишнее
    st = st.replace("'", "")[1:].replace("}", "")

print(st)

# mydict2 = {}
# mydict2 = {str(mydict)}

print(mydict)

# запись в файл
# создание каталога
if not isdir(mydir): makedirs(mydir)

with open(mydir + "/" + fake_file_nation, "w", encoding="UTF-8", newline="") as file:
    file.write(st)
    print(f"{fake_file_nation} создан")
```
